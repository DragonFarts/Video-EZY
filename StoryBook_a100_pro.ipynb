```python
!pip install pydub
from pydub import AudioSegment
import os

# Iterate over all mp3 files in the source directory
for filename in os.listdir("/content/#voicemp3"):
    if filename.endswith(".mp3"):
        # Load the MP3 file
        audio_path = os.path.join("/content/#voicemp3", filename)
        audio = AudioSegment.from_mp3(audio_path)

        # Define the duration for each segment in milliseconds (10 seconds = 10000 milliseconds)
        duration_per_segment = 10000

        # Calculate the number of segments
        num_segments = len(audio) // duration_per_segment + (1 if len(audio) % duration_per_segment else 0)

        # Create a new directory for the split files, named after the mp3 file
        new_dir = os.path.join("/content/tortoise-hf/tortoise-tts/tortoise/voices", filename.rsplit(".", 1)[0])
        os.makedirs(new_dir, exist_ok=True)

        # Split the audio and save each segment
        for i in range(num_segments):
            start_time = i * duration_per_segment
            end_time = (i+1) * duration_per_segment
            segment = audio[start_time:end_time]
            segment.export(f"{new_dir}/{i+1}.wav", format="wav")
            print(f"Saved segment {i+1}.wav for {filename}")

print("All segments saved successfully!")
```

# Load Model into gpu (may requires Restart of Runtime)

```python
import torch
import numpy as np
from diffusers import StableCascadeDecoderPipeline, StableCascadePriorPipeline
from diffusers.utils import numpy_to_pil

# Assuming the environment setup and model loading is done as per your script
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
dtype = torch.bfloat16 if torch.cuda.is_available() else torch.float32

# Initialize pipelines (ensure this matches your environment's setup)
prior_pipeline = StableCascadePriorPipeline.from_pretrained("stabilityai/stable-cascade-prior", torch_dtype=dtype).to(device)
decoder_pipeline = StableCascadeDecoderPipeline.from_pretrained("stabilityai/stable-cascade", torch_dtype=dtype).to(device)
```

```python
!rm -r /content/cascade
import random
import re
import torch
import subprocess
import os
from PIL import Image, ImageDraw, ImageFont
from moviepy.editor import ImageClip, AudioFileClip, concatenate_audioclips, AudioClip
import cv2
import numpy as np
import textwrap

!mkdir /content/cascade/

def generate_image(prompt, seed=random.randint(0, 6969), width=1024, height=1024, prior_guidance_scale=7.5, decoder_guidance_scale=0.0, num_inference_steps=50, num_images_per_prompt=1):
    # Set seed for reproducibility
    generator = torch.Generator(device=device).manual_seed(seed)

    # Generate with Prior
    prior_results = prior_pipeline(
        prompt=prompt,
        width=1024,
        height=1920,
        num_inference_steps=num_inference_steps,
        guidance_scale=prior_guidance_scale,
        generator=generator
    )

    # Extract image embeddings
    image_embeddings = None
    for result in prior_results:
        if 'image_embeddings' in result:
            image_embeddings = result['image_embeddings']
            break

    if image_embeddings is None:
        raise ValueError("Failed to generate image embeddings from the prior pipeline")

    # Generate with Decoder
    decoder_output = decoder_pipeline(
        image_embeddings=image_embeddings,
        prompt=prompt,
        num_inference_steps=num_inference_steps,
        guidance_scale=decoder_guidance_scale,
        num_images_per_prompt=1,  # This should match the decoder's expectation
        generator=generator,
        output_type="pil"
    )

    return decoder_output.images  # Adjusted to return the entire output list
# Assuming the existence of a function or script for TTS, for demonstration purposes
def generate_tts(text, output_path, voice="dan", quality="high_quality"):
    # Construct the command for TTS
    !pip install transformers==4.19.0
    !python3 /content/tortoise-hf/tortoise-tts/tortoise/do_tts.py --text '\"{text}\"' --voice="Sir David Attenborough  A message to world leaders" --candidates=1 --preset high_quality --output="/content/cascade/"
    !pip install transformers==4.37.2
    !mv "/content/cascade/Sir David Attenborough  A message to world leaders_0.wav" {tts_filename}


def add_text_to_image_opencv(image_path, text, font_size=2, font_thickness=3):
    # Load the image using OpenCV
    image = cv2.imread(image_path)

    # Choose the font
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Set the font color to white
    font_color = (255, 255, 255)

    # Wrap text to fit image width
    wrapped_text = textwrap.wrap(text, width=25)  # Adjust 'width' as needed for your image size

    # Calculate the starting y position to place text at the bottom
    # Start by calculating the total text block height (assuming all lines are evenly spaced)
    (_, text_height), _ = cv2.getTextSize(text, font, font_size, font_thickness)
    total_text_height = text_height * len(wrapped_text)

    # Calculate the starting y position
    text_y_start = image.shape[0] - total_text_height - 10  # Start 10 pixels above the bottom of the image

    for i, line in enumerate(wrapped_text):
        # Calculate text size for each line
        (text_width, text_height), _ = cv2.getTextSize(line, font, font_size, font_thickness)

        # Center the text by adjusting the starting position
        text_x = (image.shape[1] - text_width) // 2
        text_y = text_y_start + (i * text_height)  # Move down by text height for each new line

        # Add the text to the image
        cv2.putText(image, line, (text_x, text_y), font, font_size, font_color, font_thickness, cv2.LINE_AA)

    # Save the modified image
    cv2.imwrite(image_path, image)


def create_video_with_silence_and_text(image_path, audio_path, output_video_path, prompt_text, silence_duration=5):
    # Load and modify the image with text
    add_text_to_image_opencv(image_path, prompt_text, font_size=2, font_thickness=3)


    # Load the original audio
    original_audio_clip = AudioFileClip(audio_path)

    # Create a silent audio clip of the desired duration
    silence_clip = AudioClip(lambda t: [0]*2, duration=silence_duration, fps=original_audio_clip.fps)

    # Concatenate the original audio with the silent audio clip
    final_audio_clip = concatenate_audioclips([original_audio_clip, silence_clip])

    # Load the image as a clip and set its duration to the total audio duration
    image_clip = ImageClip(image_path).set_duration(final_audio_clip.duration)

    # Set the audio of the image clip to the concatenated audio clip
    final_video_clip = image_clip.set_audio(final_audio_clip)

    # Write the result to a file
    final_video_clip.write_videofile(output_video_path, fps=24)




if __name__ == "__main__":
    prompts_file_path = "/content/story.txt"

    with open(prompts_file_path, 'r') as file:
        prompts = file.readlines()

    for line_number, prompt in enumerate(prompts, start=1):
        prompt = prompt.strip()

        for iteration in range(1):
            seed = random.randint(0, 6969)
            safe_prompt = re.sub(r'[^\w\s-]', '', prompt)[:50].strip().replace(' ', '_')
            images = generate_image(prompt, seed=seed)

            for i, image in enumerate(images):
                if isinstance(image, Image.Image):
                    image_filename = f"/content/cascade/{str(line_number).zfill(5)}_{seed}_{safe_prompt}_{iteration}_{i}.png"
                    image.save(image_filename)
                    print(f"Image saved as {image_filename}")

            tts_filename = f"/content/cascade/{str(line_number).zfill(5)}_{seed}_{safe_prompt}_{iteration}_{i}.wav"
            os.makedirs(os.path.dirname(tts_filename), exist_ok=True)
            generate_tts(prompt, tts_filename)

            # Create a video for each image and corresponding TTS
            video_filename = f"/content/cascade/{str(line_number).zfill(5)}_{seed}_{safe_prompt}_{iteration}_{i}.mp4"
            create_video_with_silence_and_text(image_filename, tts_filename, video_filename, prompt, silence_duration=2)

            print(f"Video with audio and text saved as {video_filename}")
```

# Combine

```python
import os
import subprocess

# Directory containing the MP4 files
source_dir = '/content/cascade'
# Output file
output_file = '/content/final_video.mp4'

# List and sort the MP4 files alphabetically
mp4_files = sorted([os.path.join(source_dir, f) for f in os.listdir(source_dir) if f.endswith('.mp4')])

# Create a temporary file listing all the MP4 files
with open('/tmp/file_list.txt', 'w') as file_list:
    for file in mp4_files:
        file_list.write(f"file '{file}'\n")

# Use ffmpeg to concatenate the MP4 files
subprocess.run(['ffmpeg', '-f', 'concat', '-safe', '0', '-i', '/tmp/file_list.txt', '-c', 'copy', output_file])

print(f"Combined video saved as {output_file}")
```
